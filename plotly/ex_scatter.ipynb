```python
import pandas as pd
import numpy as np

import plotly
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
N = 500

trace0 = go.Scatter(
    x = np.random.randn(N),
    y = np.random.randn(N)+2,
    #name = 'Above',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'rgba(152, 0, 0, .8)',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    )
)

trace1 = go.Scatter(
    x = np.random.randn(N),
    y = np.random.randn(N)-2,
    #name = 'Below',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'rgba(255, 182, 193, .9)',
        line = dict(
            width = 2,
        )
    )
)

data = [trace0, trace1]

layout = dict(title = 'Styled Scatter',
              yaxis = dict(zeroline = False),
              xaxis = dict(zeroline = False)
             )

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='styled-scatter.html')
```

```python
df = pd.read_csv('data_plotly.csv')
df.head()
```

```python
varx = 'N'
vary = 'eps'
varname = 'E'
varhover = 'S'
varsize = 'eps'
factorsize = 10.

traces = []
for vn in df[varname].unique().tolist():
    
    dfaux = df[ df[varname] == vn ]
    
    trace0 = go.Scatter(
        x = dfaux[varx],
        y = dfaux[vary],
        name = vn,
        text = (varhover + ' : ' + dfaux[varhover].astype('str') ).values.tolist(),
        mode = 'markers',
        marker = dict(
            size = dfaux[varsize] * factorsize,
            #color = 'rgba(152, 0, 0, .8)',
            #line = dict(
                #width = 2,
                #color = 'rgb(0, 0, 0)'
            #)
        )
    )
    
    traces.append(trace0)

data = traces

layout = dict(title = 'Styled Scatter',
              yaxis = dict(title = vary,
                           zeroline = True,
                           showgrid = True,
                           showline = True),
              xaxis = dict(title = varx,
                           zeroline = True,
                           showline = True, 
                           exponentformat='e',
                           showexponent='All',
                           type = 'log')
             )

fig = dict(data= traces, layout=layout)


py.iplot(fig, filename='styled-scatter.html')
```

```python
varx = 'N'
vary = 'eps'
varname = 'E'
varhover = ['T', 'S']
varsize = 'eps'
factorsize = 20.

traces = []
for vn in df[varname].unique().tolist():
    
    dfaux = df[ df[varname] == vn ]
    
    d = pd.DataFrame()
        
    for k,i in enumerate(varhover):
        d[i] = i + ':' + dfaux[i].astype('str')
        if k < len(varhover) - 1:
            d[i] = d[i].astype('str') + ', '
    
    
    trace0 = go.Scatter(
        x = dfaux[varx],
        y = dfaux[vary],
        name = vn,
        text = d.sum(axis = 1).values.tolist(),
        mode = 'markers',
        marker = dict(
            size = dfaux[varsize] * factorsize,
            #color = 'rgba(152, 0, 0, .8)',
            #line = dict(
                #width = 2,
                #color = 'rgb(0, 0, 0)'
            #)
        )
    )
    
    traces.append(trace0)

data = traces

layout = dict(title = 'Styled Scatter',
              yaxis = dict(title = vary,
                           zeroline = True,
                           showgrid = True,
                           showline = True),
              xaxis = dict(title = varx,
                           zeroline = True,
                           showline = True, 
                           exponentformat='e',
                           showexponent='All',
                           type = 'log')
             )

fig = dict(data= traces, layout=layout)


py.iplot(fig, filename='styled-scatter.html')
```
