```python
import pandas as pd
import numpy as np

import plotly
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
df = pd.read_csv('../input/kiva_loans.csv')
df.tail()
```

```python
df = df.head(1000)
```

```python
df.columns.tolist()
```

```python
varx = 'loan_amount'
vary = 'term_in_months'
varname = 'sector'
varhover = ['country', 'activity']
varsize = 'loan_amount'
factorsize = 1./100.

traces = []
for vn in df[varname].unique().tolist():
    
    dfaux = df[ df[varname] == vn ]
    
    d = pd.DataFrame()
        
    for k,i in enumerate(varhover):
        d[i] = i + ':' + dfaux[i].astype('str')
        if k < len(varhover) - 1:
            d[i] = d[i].astype('str') + ', '
    
    
    trace0 = go.Scatter(
        x = dfaux[varx],
        y = dfaux[vary],
        name = vn,
        text = d.sum(axis = 1).values.tolist(),
        mode = 'markers',
        marker = dict(
            size = dfaux[varsize] * factorsize,
            #color = 'rgba(152, 0, 0, .8)',
            #line = dict(
                #width = 2,
                #color = 'rgb(0, 0, 0)'
            #)
        )
    )
    
    traces.append(trace0)

data = traces

layout = dict(title = 'Kiva loan amounts vs. terms in months',
              yaxis = dict(title = vary,
                           zeroline = True,
                           showgrid = True,
                           showline = True),
              xaxis = dict(title = varx,
                           zeroline = True,
                           showline = True, 
                           #exponentformat='e',
                           #showexponent='All',
                           #type = 'log'
                          )
             )

fig = dict(data= traces, layout=layout)


py.iplot(fig, filename='styled-scatter.html')
```
